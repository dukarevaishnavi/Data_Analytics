```python
# Find the sum of all values in a dictionary.
d1 = {'a':10, 'b':20, 'c':30}
print(sum(d1.values()))
# Find the maximum value in a dictionary.
print(max(d1.values()))
# Find the minimum value in a dictionary.
print(min(d1.values()))
d1 = {'a':40, 'b':60, 'c':80, 'd':30}

# Print keys whose values are greater than 50.
for key, value in d1.items():
    if value > 50:
        print(key)
        
# Print keys whose values are even numbers.
for key,value in d1.items():
    if value%2==0:
        print(key)

# Print keys whose values are odd numbers.

for key,value in d1.items():
    if value%2!=0:
        print(key)

# Replace all values with 0.
for key in d1:
    d1[key]=5
print(d1)

# Multiply all values by 2.
for key,value in d1.items():
    d1[key]=value*2
print(d1)

d1 = {}

for i in range(1,6):
    d1[i] = i*i

print(d1)
```
