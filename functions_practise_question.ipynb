```python
# Check if number is palindrome
```

```python
def reverse(n, result=0):
    if n == 0:
        return result
    return reverse(n // 10, result * 10 + n % 10)

def is_palindrome(n):
    return n == reverse(n)

print(is_palindrome(121))  # True
print(is_palindrome(123))  # False
```

```python
# Eucledian -- gcd(a,b)=gcd(b,amodb)

The Euclidean Algorithm is a method used to find the GCD (Greatest Common Divisor) of two numbers by repeatedly replacing the larger number with the remainder of the division until the remainder becomes 0.

a=int(input("enter a number_1 : "))
b=int(input("enter a number_2 : "))
while b!=0:
    a,b=b,a%b
print("GCD IS ",a)
```

```python
# Check perfect number

# A Perfect Number is a positive integer that is equal to the sum of its proper divisors (excluding the number itself).

n=int(input("enter a number : "))
sum=0
for i in range(1,n):
    if n%i==0:
        sum=sum+i
if sum==n:
    print("perfect number")
else:
    print("not a perfect number")
```
