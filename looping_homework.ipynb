```python
# 15 1. Write a while loop to print numbers from 1 to 10.
```

```python
n=1
while n<=10:
    print(n)
    n+=1
```

```python
# 17 2. Write a while loop to print all even numbers from 2 to 20.
```

```python
n=2
while n<=20:
    if n%2==0:
        print(n)
    n+=1
```

```python
# 19 3. Write a while loop to calculate the sum of numbers from 1 to 50.
```

```python
n=1
sum=0
while n<=50:
    sum=sum+n
    n+=1
print(sum)
```

```python
n = 1
while True:
    if n%3==0 and n%5==0:
        print(n)
        break
    n=n+1
```
