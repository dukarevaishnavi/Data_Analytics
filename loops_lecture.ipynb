```python
# while loop 
# runs until condition is true
# run till you have stamina and strength

# the condition is first checked
# if true --> the loop runs 
# after execution the condition is checked again
# this repeats until the condition is true when false loops stops
```

```python
i=1
while i<=5:
    print(i)
    i=i+1
```
