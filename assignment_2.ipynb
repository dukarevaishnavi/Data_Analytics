```python
Write a program to check if a given number is a Perfect Number.
A perfect number equals the sum of its divisors (excluding itself).
Example: 6 = 1 + 2 + 3 → Perfect
```

```python
n=int(input("enter a number : "))
divisor=0
for i in range(1,n):
    if n%i==0:
        divisor=i
if divisor==n:
    print(n, "is a Perfect Number")
else:
    print(n, "is NOT a Perfect Number")
```

```python
Write a program to print a pyramid pattern with numbers (n=5):
1
1 2 1
1 2 3 2 1
1 2 3 4 3 2 1
1 2 3 4 5 4 3 2 1
```

```python
n=int(input("enter a number : "))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    for k in range(i-1,0,-1):
        print(k,end=" ")
    print()
```
