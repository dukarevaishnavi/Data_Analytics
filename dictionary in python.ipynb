```python
# Count frequency of elements in a list using a dictionary.
```

```python
# Find the maximum value in a dictionary.
d1 = {"a": 1, "b": 2}
print(max(d1.values()))
```

```python
# Find the minimum value in a dictionary.
d1 = {"a": 1, "b": 2}
print(min(d1.values()))
```
