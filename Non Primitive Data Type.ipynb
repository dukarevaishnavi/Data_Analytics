```python
# Print all prime numbers in a list
```

```python
l1 = [1,2,3,4,5,6]

for i in l1:
    if i <= 1:
        continue  
    for n in range(2, i):
        if i % n == 0:
            break
    else:
        print(i)
```

```python
# Count frequency of each element in a list
```

```python
l1 = [1,2,3,4,3,4,5,6]
l2=[]
count=0
for i in l1:
    if i not in l2:
        l2.append(i)
        count=count+1
print(l2)
print(count)
```

```python
l = [1, 2, 3, 2, 4, 1, 2, 3]

freq = {}

for i in l:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print(freq)
```
