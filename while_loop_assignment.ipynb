```python
l1=[1,2,3,4,5,6,7]
target=5
start=0
end=len(l1)-1
while start<=end:
    mid=(start+end)//2
    if l1[mid]==target:
        print(f"the number is occured at {mid} index")
        break
    elif l1[mid]<target:
        start=mid+1
    else:
        end=mid-1
else:
    print("number not found")
```

```python
balance=3000
while True:
    print("1.Check Balanace")
    print("2.Deposite Money")
    print("3.Withdraw Money")
    print("4.quit")
    choice=int(input("enter your choice : "))
    if choice==1:
        print("Check balance",balance)
    elif choice==2:
        print("Deposite Money")
        deposite=int(input("Enter how much money you wnat to add : "))
        balance=balance+deposite
        print(balance)
    elif choice==3:
        print("Withdraw Money")
        withdraw=int(input("Enter how much money you want to withdraw : "))
        if withdraw<=balance:
            balance-=withdraw
            print(balance)
        else:
            print("Insufficient balance")
    else:
        print("exit")
        break
```

```python
# 87
# 33. Write a while loop to simulate a bank account system with deposit and withdrawal operations until
# user chooses to exit.
```

```python
balance=3000
while True:
    print("1.Deposite Money")
    print("2.Withdraw Money")
    print("3.Exit")
    choice=int(input("Enter Your Choice : "))
    if choice==1:
        print("Deposite Money")
        deposite=int(input("Enter how much money you want to add : "))
        balance=deposite+balance
        print("Now, Your current balance is"balance)
    elif choice==2:
        print("Withdraw Money")
        withdraw=int(input("Enter how much money you want to withdraw : "))
        if withdraw<=balance:
            balance=balance-withdraw
            print("Now, Your current balance is"balance)
        else:
            print("Insufficient Balance")
    else:
        print("Exit")
        break
```
