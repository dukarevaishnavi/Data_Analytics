```python
# Adding elements in list 

# append() - it add single element at the end of the list 
```

```python
friends=["ksbbsd","sjdbjbd","skbcsdb","wjdbcdb"]
friends.append("ebbdcewbdh")
friends.append("ebb")
print(friends)
result=["kwjeb","erkjbfb","dkjww"]
result.extend(friends)
print(result)
```

```python
# adding item at the specific location using insert()
```

```python
friends.insert(2,"efnvfvj")
print(friends)
```

```python
# removing item or element from a list
```

```python
removed=friends.pop()
print(removed)  
 # while performing pop(), first we have to store it in variable becuase pop returns removed items 
```
