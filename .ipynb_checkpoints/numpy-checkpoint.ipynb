```python
import numpy as np
arr_1=np.array([[1,2,3,4,],[2,3,4,6]])
arr_1
```

```python
# convert a ndim array i.e 2 or more than 2 d array into 1d array using flatten and ravel
```

```python
flat_arr=arr_1.flatten()
flat_arr
```

```python
flat_arr[1]=4
flat_arr
```

```python
arr_1
```

```python
# using ravel = it change the original array
```

```python
ravel_arr=arr_1.ravel()
ravel_arr
```

```python
ravel_arr[1]=4
ravel_arr
```

```python
arr_1
```

```python
z=np.random.randint(10,30,size=(3,4))
z
```

```python
flatten_arr_1=z.flatten()
flatten_arr_1
```

```python
flatten_arr_1[4]=56
flatten_arr_1
```

```python
ravel_arr_1=z.ravel()
ravel_arr_1
```

```python
ravel_arr_1[2]=90
ravel_arr_1
```

```python
z
```

```python
# what is the difference between flatten and ravel

1.flatten converts multidimensional int one dimensioanl array 
2.when make any kind of change int flatten array original array remains the same 

# where as ravel also performs similar actions but if we change item inside ravel array original array will also change 
# In simple word flatten present the view original 
# flatten is slower than ravel
# flatten also takes more memory
```
