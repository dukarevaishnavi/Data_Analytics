```python
# Check if a number is palindrome
```

```python
s = 121
n = s  
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if rev == s:
    print("Number is palindrome")
else:
    print("Number is not palindrome")
```

```python
# Print Fibonacci series up to n terms
```

```python
n = int(input("Enter number of terms: "))

n1 = 1
n2 = 1

print(n1, end=" ")
print(n2, end=" ")

for i in range(n - 2): 
    n3 = n1 + n2
    print(n3, end=" ")
    n1 = n2
    n2 = n3
```

```python
# # Find largest digit in a number-convert into int
```

```python
n = int(input("enter a number: "))

max_digit = 0

while n > 0:
    digit = n % 10
    if digit > max_digit:
        max_digit = digit
    n = n // 10

print("Largest digit is:", max_digit)
```

```python
# Count even and odd digits in a number
```

```python
n = int(input("Enter a number: "))
even_count = 0
odd_count = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    n = n // 10

print("Even digit count:", even_count)
print("Odd digit count:", odd_count)
```

```python
# Right aligned stars
```

```python
n = 4

for i in range(1, n+1):

    # spaces
    for j in range(n - i):
        print(" ", end="")

    # stars
    for k in range(i):
        print("* ", end="")

    print()
```

```python
# 10. Check Prime Numbers in List – Given a list of numbers, check which numbers are prime
# numbers. Example: nums = [3, 4, 7, 10, 11].
```

```python
numbers = [3, 4, 7, 10, 11]
for n in numbers:
    if n <= 1:
        print(n, "→ Not Prime")
    else:
        for i in range(2, n):
            if n % i == 0:
                print(n, "→ Not Prime")
                break
        else:
            print(n, "→ Prime")
```

```python
# 15. Check Number Category – Classify numbers into categories: Negative (<0), Small (0–10),
Medium (11–100), Large (>100). Example: nums = [-5, 4, 15, 120]
```

```python
nums = [-5, 4, 15, 120]
for i in nums:
    if i<0:
        print(i,"negative category")
    elif i>=0 and i<=10:
        print(i,"small category 0-10")
    elif i<=11 and i>=100:
        print(i,"medium category 11-100")
    else:
        print(i,"large category")
```

```python
# 18. Nested If – Age Category – Categorize ages as Child (<13), Teenager (13–19), Adult (20–59),
Senior (60+). Example: ages = [10, 16, 25, 65].
```

```python
ages = [10, 16, 25, 65]
for i in ages:
    if i<13:
        print("child")
    elif i<=13 and i>=19:
        print("teenager")
    elif i<=20 and i>=59:
        print("adult")
    else:
        print("senior")
```

```python
# 20. Shopping Discount Calculator – Calculate discount based on amount: ≥5000 → 20%, ≥2000 →
10%, otherwise no discount. Example: amounts = [1500, 2500, 6000].
```

```python
amounts = [1500, 2500, 6000]
for i in amounts:
    if i>=5000:
        dis = (20/100) * i
        print(i, "→ Discount:", dis)
    elif i>=2000:
        dis = (10/100) * i
        print(i, "→ Discount:", dis)
    else:
        print(i,"no discount")
```

```python
l1 = [1,2,3,4,5,6,7]

for i in l1:
    if i <= 1:
        print(i, "→ Not Prime")
    else:
        for n in range(2, i):
            if i % n == 0:
                print(i, "→ Not Prime")
                break
        else:
            print(i, "→ Prime")
```
