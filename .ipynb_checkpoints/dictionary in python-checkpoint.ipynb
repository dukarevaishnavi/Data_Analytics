# create dictionary uisng zip method
# for zip method we declare list of keys different and list of values different

```python
k1=["company","founder","valuation"]
k2=["google","larry paige","Three Trillion Dollar"]
company=dict(zip(k1,k2))
company
```

```python
type(company)
```
