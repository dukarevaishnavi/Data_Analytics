# Membership Operator = checks an item present inside the object 
# 1. IN - Chekcks an item present inside the object returns true if item is present and false when item is not 
# 2. NOT IN - Chekcks an item present inside the object returns true if item is not present and false when item is present

```python
l1=["apple","eagle","rose","lion"]
print("apple" in l1)
```

```python
print("Apple" in l1)
```

```python
t1=("apple","eagle","rose","lion")
print("apple" in t1)
```

```python
s1="machine learning "
print("m" in s1)
```

```python
print("learn" in s1)
```

```python
d1={"name":"vaishnavi","age":21,"city":"solapur"}
print(21 in d1.values())
```

```python
# NOT IN 
```

```python
t1=("apple","eagle","rose","lion")
print("apple" not in t1)
```

```python
d1={"name":"vaishnavi","age":21,"city":"solapur"}
print(21 not in d1.values())
```

```python
t1=("a","e","r","l")
print("t" not in t1)
```

# Identity Operator = check if both objects that are getting compared are the one object or not 
# IS - returns true if both variable points towards exactly same object and returns false if both object sits at different memiry 
# IS NOT - Works exactly opposite to IS
# '==' compares if both object contains same items or not.

```python
# IS Operator 
```

```python
a="hello"
b="hello"
print(a is b)
```

```python
c="Python"
d=c
print(c is d)
```

```python
print(c is d)
```

```python
x=[1,2]
y=x
print(x is y)
```

```python
s1="i like to read books"
s2="i like to read books"
print(s1 is s2)
```

```python
s1="i like to read books"
s2="i like to read books"
print(s1 is not s2)
```

```python
s3="abc"
s4="abc"
print(s3 is not s4)
```

```python
x=[1,2]
y=x
print(x is not y)
```

```python
d1={"name":"vaishnavi","age":21,"city":"abcdef"}
d2={"name":"vaishnavi","age":21,"city":"abcdef"}
print(d1 is not d2)
```
