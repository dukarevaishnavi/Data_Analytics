```python
# Part 2: Data Loading & Exploration (20 points)


# 6. Load the sales_data.csv file into a Pandas DataFrame

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sales_data=pd.read_csv(r"C:\Users\HP\Downloads\sales_data.csv")
sales_data
```

```python
# 7. Display the first 5 rows and basic information about the dataset
sales_data.head()
```

```python
# 8. Check for missing values and report the count in each column
sales_data.isnull().sum()
```

```python
# 9. Identify and report the data type of each column
sales_data.dtypes
```

```python
# 10. Calculate the total number of records and unique products in the dataset
sales_data.count()
len(sales_data)
sales_data["product"].unique()
```

```python
# 11. Handle missing values in the 'discount' column by replacing with the median value
sales_data["discount"].fillna(sales_data["discount"].median())
```

```python
# 12. Convert the 'date' column to datetime format
sales_data["date"]=pd.to_datetime(sales_data["date"])
sales_data
```

```python
# 13. Create a new column 'month' extracted from the date
sales_data["month"]=sales_data["date"].dt.month
sales_data
```

```python
# 14. Remove any duplicate rows
sales_data.duplicated()
sales_data.drop_duplicates()
```

```python
# 15. Create a 'revenue' column calculated as: quantity × price × (1 - discount)

sales_data["revenue"]=(sales_data["quantity"]*sales_data["price"])*(1-sales_data["discount"])
sales_data
```

```python
# Calculate and display the following statistics:
# 16. Summary statistics (mean, median, std) for 'quantity', 'price', and 'revenue'
sales_data[["quantity","price","revenue"]].agg(["mean","median","std"])
```

```python
# 17. Total revenue and average transaction value
sales_data["revenue"].sum()
```

```python
sales_data["revenue"].mean()
```

```python
# 18. Top 5 products by total revenue
sales_data["product"].sort_values().head()
```

```python
# 19. Sales by category (count and total revenue)
sales_data.groupby("category").agg({
    "category":"count",
    "revenue":"sum"
})
```

```python
# 20. Monthly revenue trend (total revenue for each month)
sales_data.groupby("month")["revenue"].sum()
```

```python
# 22. Histogram: Distribution of product prices
plt.hist(sales_data["product"],bins=15,edgecolor="red",color="pink")
plt.title("Distribution of product prices")
plt.xlabel("product")
plt.ylabel("product_price")
plt.show()
```

```python
# 23. Bar chart: Revenue by category
plt.bar(sales_data["revenue"],sales_data["category"])
plt.xlabel("category")
plt.ylabel("revenue")
plt.show()
```

```python
# 25. Scatter plot: Relationship between quantity and revenue
plt.scatter(sales_data["quantity"],sales_data["revenue"])
plt.title(" Relationship between quantity and revenue")
plt.xlabel("quantity")
plt.ylabel("revenue")
plt.show()
```

```python
# 24. Line plot: Monthly revenue trend
plt.plot(sales_data["month"],sales_data["revenue"],color="pink")
plt.title("Monthly revenue trend")
plt.xlabel("month")
plt.ylabel("revenue")
plt.show()
```
