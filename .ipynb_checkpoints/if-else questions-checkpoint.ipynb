```python
# Write a program to check whether a given character is a vowel or not.
```

```python
s = input("enter a string : ")

if ('a' in s[0:] or 
    'e' in s[0:] or
    'i' in s[0:] or
    'o' in s[0:] or
    'u' in s[0:]):
    print("string contains vowels")
else:
    print("string not contains vowels"
```

```python
# Take three numbers and print the largest using if-else statements.
```

```python
n1=int(input("enter a number_1 :"))
n2=int(input("enter a number_2 :"))
n3=int(input("enter a number_3 :"))
if n1>n2 and n1>n3:
    print(f"{n1} is greater")
elif n2>n3 and n2>n1:
    print(f"{n2} is greater")
else:
    print(f"{n3} is greater")
```
