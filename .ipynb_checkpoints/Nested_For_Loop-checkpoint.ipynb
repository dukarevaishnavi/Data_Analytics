```python
for i in range(outer_range):
    for j in range(inner_range):
       # code block
```

```python
for i in range(3):
    for j in range(4):
        print(i,j)
```

```python
for i in range(1,6):
    print("this is outer loop")
    for j in range(1,5):
        print("this is inner loop")
    print()
```

```python
for i in range(3):
    for j in range(4):
        print("i=",i,"j=",j)
```

```python
for i in range(1,5):
    for j in range(i):
        print("*",end=" ")
    print()
```

```python
for i in range(4):
    for j in range(5):
        print("hello world")
# how many times the loop will run
```

```python
# let's print stars
```

```python
count=1
for i in range(3):
    for j in range(3):
        print("*",end=" ")
        count+=1
    print(count)
    
```

```python
count=1
for i in range(9):
    for j in range(11):
        print("*",end=" ")
        count+=1
print(count)
```

```python
n=4
for i in range(1,16):
    for j in range(1,16):
        print(i*j,end="\t")
    print()
# i controls the number of rows 
# j controls the multiplication
```

```python
for i in range(1,21):
    for j in range(21,31):
        print(j*i,end="\t")
    print()
```
