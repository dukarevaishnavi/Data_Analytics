````python
# multiple keyword arguments using double star ** kwargs and it is stored as dictionary 
````

````python
def info(**data):
    print(data)
info(name="vaishnavi",age=26,college="wkqbsdkwjdbk",hobby="chess")
````

````python
def info(**data):
    for i in data:
        return data
info(name="vaishnavi",age=26,college="wkqbsdkwjdbk",hobby="chess")
````

````python
def details(**info):
    for key,value in info.items():
        print(key,value)
details(name="vaishnavi",age=26,college="wkqbsdkwjdbk",hobby="chess")
````

````python
# Recursion - Function calling itself. It is used to solve smaller problems of same version .
# in consists of 2 parts .
# base case(stopping condition)
# it prevents infinite loop 
# function stops calling itself when criteria meets.

# 2. Recursive case 
# the function call itself with smaller input. This helps to reduce bigger problem and to finally reach at stopping condition.
````

````python
Syntax :
def function():
    if base_case:
        return something
    return function(smaller_problem)
````

````python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
fact(5)
````

````python
def factorial(n):
    if n==0:
        return 1
    return n*factorial(n-1)  # reduce
factorial(5)
````
