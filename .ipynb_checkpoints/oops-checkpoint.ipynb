```python
# create a class student 
# take attributes name surname city age and dob
# /craete 3 methods 
# info() will return its name and surname
# age will return age and dob
class Student:
    def __init__(self,name,surname,city,age,dob):
        self.name=name
        self.surname=surname
        self.city=city
        self.age=age
        self.dob=dob
    def info(self):
        print(f"my name is {self.name} and my surname is {self.surname}")
    def show_age(self):
        print(f"my age is {self.age} and my dob is {self.dob}")
    def location(self):
        print(f"location of {self.name} is {self.city}")
c4=Student("vaishnavi","dukare","pune",21,"21.1.2005")
```

```python
c4.show_age()
```

```python
c4.info()
```

```python
c4.location()
```

```python
# calculator class 
# add,divide,substrct,multiply,square and cube
class Calculator:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def add(self):
        print(f"The addition of a and b is {self.a+self.b}")
    def substract(self):
        print(f"The substraction of a and b is {self.a-self.b}")
    def division(self):
        try:
            c=a.self/b.self
            print(c)
        except:
            print("invalid value")
    def multiply(self):
        print(f"The multiplication of a and b is {self.a*self.b}")
    def square(self):
        print(f"the square of a is {self.a*self.a} \n the square of b is {self.b*self.b}")
    def cube_1(self):
        try:
            if b==0 or a==0:
                print(f"{self.a**self.b}=1")
        except:
            print("invalid numbers")

c5=Calculator(4,0)
```

```python
c5.cube_1()
```

```python
class BankAccount: 
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance # amking this balance private /hinding
    def deposite(self,amount):
        self.__balance+=amount
        print(f"{amount} depsoited in account .New Balance :{self.__balance}")
    def withdraw(self,amount):
        self.__balance-=amount
        print(f"{amount} withdrawn from account .New Balance :{self.__balance}")
    def check_balance(self):
        print(f"current balance is {self.balance},account_holder:{self.name}")

acc_1=BankAccount("Vaishnavi",10000)
```

```python
acc_1.balance
```

```python
# this process is known as encapsulation. It protects the data 
```

```python
acc_1.balance=837476
acc_1.check_balance()
```

```python
# parent class
class Animal:
    def eat(self):
        print("animals eat food")
    def sleep(self):
        print("animals sleeps at night")
# child class
class Dog(Animal):
    def bark(self):
        print("Dog barks louldy")
class Cat(Animal):
    def mewow(self):
        print("cat mews")
d=Dog()
c=Cat()

# inheritance- child class inherits the methods from parent class
```

```python
d.bark()
```

```python
d.sleep()
```

```python
d.eat()
```

```python
c.eat()
```

```python
class Student:
    def role(self):
        print("student study")
class Teacher(Student):
    def role(self):
        print("Teacher Teaches")
t=Teacher()
s=Student()
```

```python
t.role()
```

```python
s.role()
```

```python
# single method "role()" has multiple forms across classes
# means single method performs different task
# knawn as polymorphism
```

```python
# to perform abststraction we require library
from abc import ABC,abstractmethod 
# this means this is abstract class
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
# child class implementation 
class Car(Vehicle):
    def start(self):
        print("car starts with key Fob")
# second child class 
class Bike(Vehicle):
    def start(self):
        print("bike starts with side kick")
class Cycle(Vehicle):
    def start(self):
        print("cycle")
c=Car()
b=Bike()
c1=Cycle()
```

```python
c.start()
```

```python
b.start()
```

```python
c1.start()
```
