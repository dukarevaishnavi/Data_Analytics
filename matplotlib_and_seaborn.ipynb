```python
# Correlation
# --shows the relationship bteween the two entities 
# 1.positive correlation - one gets increases other also increase
# 2.negative correlation - one gets increased other gets decreased 

# correlation has a range 
# -(-1,0,1)
# -1 to 0 is scale of negative correlation
# 0 to +1 is scale of positive correlation 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
tips=sns.load_dataset("tips")
tips
```

```python
corr=tips.corr(numeric_only=True)
```

```python
corr
```

```python
sns.heatmap(corr,cmap="coolwarm",annot=True)
plt.show()
```

```python
# 0.1 to 0.4 - positive relationship is there but weak positive
# 0.4 to 0.7 - strong relationship
# 0.7 to 1 or 0.99 - very strong relationship
```

```python
x=np.linspace(0,2*np.pi,100)
y=np.cos(x)
plt.plot(x,y)
plt.title("cosine wave")
plt.show()
```

```python
# Display both sine and cosine in same graph.
# Save the plot as PNG.
x=np.linspace(0,2*np.pi,100)
y_cos=np.cos(x)
y_sin=np.sin(x)
plt.figure(figsize=(3,2))
plt.plot(x,y_sin,label="sine wave")
plt.plot(x,y,label="cos wave")
plt.legend()
plt.savefig("sine_cosine graph")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

y1 = [10, 20, 15, 25]
y2 = [5, 15, 10, 20]

fig, ax = plt.subplots(2, 1)  # 2 rows, 1 column

ax[0].plot(x, y1)
ax[0].set_title("Plot 1")

ax[1].plot(x, y2)
ax[1].set_title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
# Plot different graphs in each subplot.
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,1000,25)
fig,axs=plt.subplots(2,2,figsize=(10,8))
# line plot 1
axs[0,0].plot(x,np.sin(x),color="blue")
axs[0, 0].set_title("Line Plot (Sine)")

# line plot 2
axs[0,1].scatter(x,np.cos(x),color="red")
axs[0,1].set_title("Line Plot (Cosine)")

# plot 3 - Bar Chart
categories=["A","B","C","D","E"]
values = [15, 25, 10, 20,25]
axs[1,0].bar(categories,values,color="orange")
axs[1,0].set_title("bar chart")

# histogram plot 4 - Histogram
data=np.random.randn(1000)
axs[1,1].hist(data, bins=20, color="green")
axs[1,1].set_title("histogram")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
fig,axs=plt.subplots(2,2,figsize=(10,8),sharex=True)


# plot 1
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title("Sine")

# cosine

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("cos")

# tangent

axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("Tangent")


# linear

axs[1,1].plot(x,x)
axs[1,1].set_title("linear")

plt.tight_layout()
plt.show()
```
