# Set Operations 

```python
a={1,2,3,4,5,6}
b={7,8,9,1,2,3}
print(a,b)
```

```python
a.union(b)  #it combines all element but is removes the duplicates
a.intersection(b) # it gives common elements

# using pipe symbol

a|b
a&b
```

```python
# difference - returns elements from set b that are different from set a

a-b
a.difference(b)
b.difference(a)
```

```python
# key takeaway a-b!=b-a
```

```python
# symmetric difference

a.symmetric_difference(b)
a^b
```

```python
# Subset - set a is a subset of set b if all elements in set a contains the value of set b
```

```python
s1={1,2,3}
s2={1,2,3,4,5}
print(s1,s2)
print(s1.issubset(s2))
```

```python
# using operator <= 
s1<=s2
```

```python
set_1={"mango","banana","peru"}
set_2={"cherry","mango","banana","peru","watermelon"}
set_1<=set_2
```

# Superset - set A is superset of B is A contains all elements of B 

```python
a={1,2,3,4,5,6}
b={1,2,3}
a.issuperset(b)
```

```python
b.issuperset(a)
```

```python
b.issubset(a)
```

```python
# A is superset of B is since all elements from b is present in A
# B is subset of A since all elements in B are Present in A
```

```python
a>=b
```
