```python
# What is normalization?

Normalization means scaling your data to a standard range, usually 0 to 1, so it’s easier to compare values across different columns or datasets.

    normalized=max(x)−min(x)/x−min(x)
```

```python
# Normalize passengers
flights["passengers"]=flights["passengers"]-flights["passengers"].min()/flights["passengers"].max()-flights["passengers"].min()
```

```python
A pivot table is a way to summarize data in a table format.

You can aggregate values (like sum, mean, count) based on rows and columns.
```

```python
# Create pivot table (year vs month)
pivot=flights.pivot_table(index="year", 
    columns="month", 
    values="passengers", 
    aggfunc="mean"  )
pivot
```

```python
# it shows frequencies or aggregated values for combinations of two (or more) categorical variables. 😎
What is a crosstab?
pd.crosstab() creates a table of counts (or sums) of two categorical variables.
Rows = one category, Columns = another category
Can also include aggregation (values + aggfunc)

import pandas as pd

pd.crosstab(flights["year"], flights["month"])
```
