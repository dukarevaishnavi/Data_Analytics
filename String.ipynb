# Slicing - It has start index and stop index 
. Start index  - from which item, should make a slice .
. Stop index - till which item we should stop a slice and that item is exclusive .

```python
s1="i like python"
print(s1[:6]);
```

```python
# Slicing also has Stepping 
# [Start index : stop index : step]
# step simply means after how much steps we are taking to select assign
```

```python
s1="i like jdbsbcjdcb"
print(s1[0:6:2])
```

```python
s1="retupmoc"
print(s1[7:-5:-1])
```

```python
# Replace all vowels with *
```

```python
s="hello"
for i in s:
    if i in "aeiou":
        s=s.replace(i,"*")
print(s)
```

```python
# Count words in a string
```

```python
s1="hello world python"
s2=s1.split()
print(s2)
count=0
for i in s2:
    count=count+1
print(count)
```

```python
# Check if two strings are anagrams
```

```python
s1="listen"
s2="silent"
if sorted(s1)==sorted(s2):
    print("anagram")
else:
    print("not anagram")
```

```python
# Find duplicate characters
```

```python
s="aabbbccc"
for i in s:
    if s.count(i)>1:
        print(i)
```
