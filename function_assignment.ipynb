```python
# Write a function count_vowels(s) that returns the number of vowels in a string.
```

```python
def count_vowels(s):
    count=0
    for i in s:
        if i in "aeiou":
            count=count+1
    print(count)
count_vowels("abc")
```

```python
# Write a function is_palindrome(s) that returns True if the string is a palindrome.
```

```python
def is_palindrome(s):
    rev=""
    for i in s:
        rev=i+rev
    print(rev)
    if rev==s:
        print("yes palindrome")
    else:
        print("not a palindrome")
is_palindrome("madam")  
```

```python
# Write a function sum_list(lst) that takes a list of numbers and returns their sum without using Python's
# built-in sum().
```

```python
def sum_list(lst):
    sum=0
    for i in lst:
        sum=sum+i
    print(sum)
sum_list([1,2,3,4,5])
```

```python
# Write a function using *args called total(*numbers) that returns the sum of any number of arguments.
```

```python
def total(*numbers):
    total=0
    for i in numbers:
        total=total+i
    print(total)

total(4,5,6,4,3,9)    
```

```python
# Write a function using **kwargs called display_info(**details) that prints each key-value pair.
```

```python
def display_info(**details):
    for keys,values in details.items():
        print(keys,":",values)
display_info(name="vaishnavi",age=45,college="a.g patil instituet of technology")
```
