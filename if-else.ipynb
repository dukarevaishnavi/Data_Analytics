```python
# create grading system using if elif else
marks=int(input("enter your marks : "))
if marks >=90:
    print("A+ GRADE")
elif marks>=80:
    print("A GRADE")
elif marks>=70:
    print("B GARDE")
elif marks>=60:
    print("C GRADE")
elif marks>=50:
    print("D GRADE")
else:
    print("fail")
```

```python
a = int(input("enter first number: "))
b = int(input("enter second number: "))
op = int(input("enter an operator: "))

if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
elif op == "//":
    print(a // b)
else:
    print("invalid operator")
```
